```python
def print_memories(memories: list[Document]) -> None:
    for i, memory in enumerate(memories):
        print("-" * 100)
        print(f"Memory {i + 1}:")
        print(f"{i + 1}. {memory.page_content[:100]}")
        print(f"Source: {memory.metadata['source']}")
        print("-" * 100)
```
